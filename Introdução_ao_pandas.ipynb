Series

Uma Series é um objeto de tipo array unidimensional que contém uma
sequência de valores (de tipos semelhantes aos tipos do NumPy) de mesmo
tipo e um array associado com rótulos de dados chamado índice


```python
import pandas as pd
import numpy as np
```

```python
#declarando uma Series
obj =pd.Series([4,7,-5,3])
obj
```

Atributos de Series

```python
#atributos
obj.array
```

```python
obj.index
```

```python
#Series com rótulos
obj2 = pd.Series([4,7,-5,3], index=['d', 'b', 'a', 'c'])
obj2
```

```python
obj2.index
```

```python
obj2["a"]
```

```python
obj2[obj2 > 0]
```

```python
obj2 * 2
```

```python
import numpy as np
np.exp(obj2)
```

```python
"b" in obj2
```

```python
"e" in obj2
```

```python
#Criando uma serie pelo dicionario
sdata = {"Ohio":3500, "Texas":7100, "Oregon":1600, "Utah":5000}
obj3 = pd.Series(sdata)
obj3
```

```python
#Conversão de serie em dicionario
obj3.to_dict()
```

```python
#sobre escrevedno Keys do dicionario
states = ["California", "Ohio", "Oregon", "Texas"]
obj4 = pd.Series(sdata, index=states)
obj4
```

```python
#valores ausestes,nulos
pd.isna(obj4)
```

```python
pd.notna(obj4)
```

```python
#alinhamento automático pelo rótulo do índice em aritméticas
obj3 + obj4
```

```python
#colocando nome na coluna
obj4.name = "population"
obj4.index.name = "state"
obj4
```

```python
#indice por atribuição
obj
obj.index = ["Bob", "Steve", "Jeff", "Ryan"]
obj
```

Data Frame

Um DataFrame representa uma tabela de dados retangular e contém uma
coleção ordenada e nomeada de colunas, cada uma podendo ter um tipo de
valor diferente (numérico, string, booleano etc.).

```python
#criando um dara frame atras de um dicionario de listas
data = {"state":["Ohio","Ohio", "Ohio","Nevada", "Nevada", "Nevada"],
        "year":[2000,2001,2002,2001, 2002, 2003],
        "pop": [1.5,1.7,3.6,2.4,2.9,3.2]}
frame = pd.DataFrame(data)
frame
```

```python
# mostra incio e fim
x =frame.head()
print(x)
print("_______")
y = frame.tail()
print(y)
```

```python
#colunas especificadas
pd.DataFrame(data, columns=["year", "state", "pop"])
```

```python
frame2 = pd.DataFrame(data, columns=["year", "state", "pop", "debt"])
frame2
```

```python
#recuperação de series uma coluna dataframe
frame2["state"]
```

```python
frame2.year
```

```python
#acesso a atributos
frame2.loc[1] #eixo
```

```python
frame2.iloc[2] #inteiros
```

```python
#atribuição
frame2["debt"] = 16.5
frame2
```

```python
frame2["eastern"] = frame2["state"] == "Ohio"
frame2
```

```python
populations = {"Ohio": {2000: 1.5, 2001: 1.7, 2002: 3.6},
               "Nevada": {2001: 2.4, 2002: 2.9}}
populations
```

```python
frame3 = pd.DataFrame(populations)
frame3
```

```python
#transposição, troca linhas e colunas
frame3.T
```

```python
#retorno bidimensional
frame3.to_numpy()
```

Objetos Index

Os objetos Index do pandas são responsáveis por conter os rótulos dos eixos
(incluindo o nome das colunas de um DataFrame) e outros metadados (como
o nome ou os nomes dos eixos).

```python
from numpy._core.defchararray import index
obj = pd.Series(np.arange(3), index=["a", "b", "c"])
obj
```

```python
index = obj.index
index
```

```python
index[1:]
```

```python
#objetos index são imutaveis
index[1] =  'd'
```
